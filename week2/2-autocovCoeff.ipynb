```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import acf
import pylab
plt.rcParams["figure.figsize"] = (12,8)
```

# Autocovariance Coefficients

```python
# Example of white noise or random process

# Define a random seed to make our data reproducible
np.random.seed(45)

# generate a purely random process also known as white noise, where:
# loc is mean
# scale is standard deviation
white_noise = np.random.normal(loc=0, scale=2, size=1000)

plt.plot(white_noise);
```

```python
# let's check the mean and the standard deviation
# Multiple lines f printing
print(f'mean: {white_noise.mean()} \n'
        f'standard deviation: {white_noise.std()}')
```

```python
# plot a white noise histogram 
plt.hist(white_noise, bins=21);
```

```python
# Q-Q plot of the white nois
stats.probplot(white_noise, dist='norm', plot=pylab);
```

From the two above pictures we can conclude that the white noise is normally distributed.   

The white noise is a collection of uncorrelated random variables $w_t$, with mean 0 and finite variance $\sigma_w^2$ denoted as $w_t \text{ ~ }wn(0, \sigma_w^2)$.   
We will sometimes require the noise to be independent and indetically distributed (iid) random variables with mean 0 and variance $\sigma_w^2$. We distinguish this by writing  $w_t ~\text{~}~iid(0, \sigma_w^2)$ or white indepent noise, iid noise.   

The designation white originates from the analogy with white light and indicates that all possible periodic oscillations are present with equal strength.

```python
# let's plot the autocorrelation of the white noise.
# to avoid getting twice the same plot we assign the result of the plot_acf to a 
# ranodm variable - in this case
p_acf = plot_acf(white_noise, lags=22)
```

As we can see the white noise doesn't corrolate with time-shifted-self.   
The blue area shows us the significance level. By default, this is set to a 95% confidence interval, suggesting that correlation values outside of this cone are very likely a correlation and not a statistical fluke.

So this plot tells us that there are no lags in the previous steps.

```python
# Acf coefficients
acf(white_noise, nlags=22, fft=False)
```

All the lags have close to zero autocorrelation coefficients. This means that all the lagged variables are purely random and not useful for prediction.
