## Stationarity

Three stationarity propterties:
* No systematic change in mean, i.e. no trend
* No systematic change in variation, like in jj
* No periodic fluctuations

Stationary time series don't have any systematic change in the variation nor periodic fluctuations.  
So basically, the properties of one section of a data in a time series are much like the properties of the other sections of the data.  Which is almost impossible to find in time series. 

Stationarity is a property of stochastic process of a model, not a time series. However, to be able to model a time series we need first to transform it to a stationary state.

What we do, we basically do some transformations to get the stationary time series. Once we have a stationary time series, we model it and then we go back and we model our non-stationary time series. So we use the transformations as a middle step.
